- 지지도(support) : 한 거래 항목 안에 A와 B를 동시에 포함하는 거래의 비율. 지지도는 A와 B가 함께 등장할 확률이다. 전체 거래의 수를 A와 B가 동시에 포함된 거래수를 나눠주면 구할 수 있다.
- 신뢰도(confidence) : 항목 A가 포함하는 거래에 A와 B가 같이포함될 확률. 신뢰도는 조건부 확률과 유사하다. A가 일어났을 때 B의 확률이다. A의 확률을 A와 B가 동시에 포함될 확률을 나눠주면 구할 수 있다.
- 향상도(lift) : A가 주어지지 않을 때의 품목 B의 확률에 비해 A가 주어졌을 때 품목 B의 증가 비율. B의 확률이 A가 일어났을 때 B의 확률을 나눴을 때 구할 수 있다. lift 값은 1이면 서로 독립적인 관계이며1보다 크면 두 품목이 서로 양의 상관관계, 1보다 작으면 두 품목이 서로 음의 상관관계이다. A와 B가 독립이면 분모, 분자가 같기 때문에 1이 나온다.

### 1. 20대 연관분석

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
ap_20 = []
for i in range(age_20.shape[0]):
    ap_20.append(list(age_20.loc[i,"숙박":][age_20.loc[i,"숙박":].notnull()].index))

te = TransactionEncoder()
te_ary = te.fit(ap_20).transform(ap_20)
df_ap_20 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df_ap_20, min_support=0.05, use_colnames=True)
frequent_itemsets.sort_values(by="support", ascending=False)[:20]
```

```python
association_rules(frequent_itemsets, metric="lift", min_threshold=0).sort_values(by="lift", ascending=False)[:20]
```

```python
association_rules(frequent_itemsets, metric="lift", min_threshold=0).sort_values(by="lift", ascending=True)[:20]
```

- 소비재, 식음료, 온라인 결제 - 지지도 80% 넘는다.
- 거래의 비율이 세 항목 너무 많이 차지하고 있으므로, 다른 열들의 연관관계를 파악하기 위해 제거해준다. 

#### 소비재, 식음료, 온라인결제 제거

```python
age_20_drop = age_20.drop(["소비재","식음료", "온라인결제"], axis=1)
```

```python
ap_20_drop = []
for i in range(age_20_drop.shape[0]):
    ap_20_drop.append(list(age_20_drop.loc[i,"숙박":][age_20_drop.loc[i,"숙박":].notnull()].index))

te = TransactionEncoder()
te_ary = te.fit(ap_20_drop).transform(ap_20_drop)
df_ap_20_drop = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df_ap_20_drop, min_support=0.05, use_colnames=True)
frequent_itemsets.sort_values(by="support", ascending=False)[:20]
```

```python
association_rules(frequent_itemsets, metric="lift", min_threshold=0).sort_values(by="lift", ascending=False)[:15]
```

- lift(향상도) 기준 내림차순 배열한 결과이다. 
- 차량, 의료계, 문화생활, 유류 간의 양의 상관관계가 존재함을 알 수 있다. 

```python
association_rules(frequent_itemsets, metric="lift", min_threshold=0).sort_values(by="lift", ascending=True)[:15]
```

- 향상도 기준 오름차순 배열한 결과이다. 
- 이동수단과 유류는 음의 상관관계를 가진다. => 지출을 이동수단에 할 수록 유류에는 하지 않는다는 의미로 해석된다. 
- 마찬가지로 차량과 유류는 이동수단과 음의 상관관계를 가진다. 
  => 차량에 필요한 지출 = 차량을 소유하고 있다는 의미 => 이동수단에 필요한 대중교통 이용 확률 적어짐
