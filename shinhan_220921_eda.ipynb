```python
#ANOVA. 나이대별 결제금액 차이가 존재할 것이다
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
model = ols('B167 ~ age', df).fit()
anova_lm(model)
```

```python
#Bonferoni. 어떤 나이대별 차이가 유의한지 사후검정
from statsmodels.sandbox.stats.multicomp import MultiComparison
import scipy.stats
comp = MultiComparison(df.B167, df.age)
result = comp.allpairtest(scipy.stats.ttest_ind, method='bonf')
result[0]
```

```python
#T-Test. MZ vs YOLD 결제금액 차이가 존재할 것이다
def make_label(x):
  if x in ['20대', '30대']:
    return 'MZ'
  else:
    return 'YOLD'

def do_ttest(df, col):
    df['label'] = df['age'].apply(make_label)
    df_A = df[df['label']=='MZ']
    df_B = df[df['label']=='YOLD']
    sns.distplot(df_A[col], color='red') 
    sns.distplot(df_B[col], color='blue')
    from scipy import stats
    t_stat, p_value = stats.ttest_ind(df_A[col], df_B[col], equal_var=True) 
    print('t-통계량:', t_stat, 'p-값:', p_value)

do_ttest(df, 'B167')
```
